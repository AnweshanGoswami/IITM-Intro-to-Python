Accept a word as input and do the following:
 (1) If it is a perfect word, print It is a perfect word
 (2) If the word is not a perfect word, print It is not a perfect word

```python
word = input().lower()
flag1 = True
flag2 = False
flag3 = False

vowels = "aeiou"
for i in vowels:
    if i not in word:
        flag1 = False
        break

if flag1:
    a = word.index("a")
    e = word.index("e")
    i = word.index("i")
    o = word.index("o")
    u = word.index("u")
    if a < e < i < o < u:
        flag2 = True

ac=ec=ic=oc=uc=0
for x in word:
    if x == "a":
        ac += 1
    elif x == "e":
        ec += 1
    elif x == "i":
        ic += 1
    elif x == "o":
        oc += 1
    elif x == "u":
        uc += 1
if ac>=ec>=ic>=oc>=uc:
    flag3 = True

if flag1 and flag2 and flag3:
    print("It is a perfect word")
else:
    print("It is not a perfect word")

```
