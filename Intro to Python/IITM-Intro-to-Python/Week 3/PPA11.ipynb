```python
n = int(input())
trips = list()

for i in range(n):
    for j in range(1, i+1):
        for k in range(1,j):
            if (i**2) == (j**2) + (k**2):
                trips.append([k,j,i])
trips.sort()
if len(trips) == 0:
    print("NO SOLUTION")
else:
    for a in trips:
        for b in range(len(a)-1):
            print(a[b], end= ",")
        print(a[2])
```
